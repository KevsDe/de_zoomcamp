SELECT  t2."Borough", SUM(t1.total_amount) AS total_amount
FROM green_taxi_trips AS t1
INNER JOIN taxi_zone AS t2
ON t1."PULocationID" = t2."LocationID"
WHERE lpep_pickup_datetime >= '2019-09-18' AND lpep_pickup_datetime < '2019-09-19'
AND t2."Borough" <> 'Unknown'
GROUP BY 1
HAVING SUM(t1.total_amount) > 50000
;


SELECT t3."Zone", MAX(tip_amount)
FROM green_taxi_trips AS t1
INNER JOIN taxi_zone AS t2
ON t1."PULocationID" = t2."LocationID"
INNER JOIN taxi_zone AS t3
ON t1."DOLocationID" = t3."LocationID"
WHERE lpep_pickup_datetime >= '2019-01-01' AND lpep_pickup_datetime < '2019-10-01'
AND t2."Zone" = 'Astoria'
GROUP BY t3."Zone"
ORDER BY 2 DESC
;
